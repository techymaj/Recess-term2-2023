# Assignemt 1: Series with days & dates from a dictionary

```python
import pandas as pd
dataset = {
    'days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'dates': ['01-07-2023', '02-07-2023', '03-07-2023', '04-07-2023', '05-07-2023', '06-07-2023', '07-07-2023']
}
s = pd.Series(dataset)
days = []
index = -1
for j in s['days']:
    index = index + 1
    for i in s['dates']:
        days.append(i)
    print(j,":", days[index])
```

# Assignment 2: Create a series using only data from Monday and Tuesday

```python
import pandas as pd
dataset = {
    'days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'dates': ['01-07-2023', '02-07-2023', '03-07-2023', '04-07-2023', '05-07-2023', '06-07-2023', '07-07-2023']
}
new_dataset = []
for day in dataset['days']:
    if day == 'Monday' or day == 'Tuesday':
        new_dataset.append(day)
s = pd.Series(new_dataset)
print(s)
```

### Also use the appropriate methods to remove:

- ##### Wrong data & format for the Work.csv file

```python
# Removes rows with a py-score that is not an int
work_df = work_df[pd.to_numeric(work_df['py-score'], errors='coerce').notnull()] 
work_df
```

```python
# Removes rows with age that is not an int
work_df = work_df[pd.to_numeric(work_df['age'], errors='coerce').notnull()] 
work_df
```

- ##### Wrong data & format for the mine.csv file

```python
# Removes rows with Duration, Pulse, Maxpulse & Calories that is not an int. Reduces rows from 169 - 164
mine_df = mine_df.loc[pd.to_numeric(mine_df['Duration'], errors='coerce').notnull()] 
mine_df = mine_df.loc[pd.to_numeric(mine_df['Pulse'], errors='coerce').notnull()] 
mine_df = mine_df.loc[pd.to_numeric(mine_df['Maxpulse'], errors='coerce').notnull()] 
mine_df = mine_df.loc[pd.to_numeric(mine_df['Calories'], errors='coerce').notnull()] 
mine_df
```
